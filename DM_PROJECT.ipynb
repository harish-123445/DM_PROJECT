## K-Means Clustering

```python
# Importing Required Libraries
from sklearn.cluster import KMeans
from sklearn.preprocessing import StandardScaler
from sklearn.metrics.pairwise import cosine_similarity
```

```python
# Extracting the movie title and features
movie_titles = data['title']
movie_features = data.drop(columns=['title'])
```

```python
# Scaling the features
scaler = StandardScaler()
movie_features_scaled = scaler.fit_transform(movie_features)
```

```python
# Determining the best k-values using elbow method
wcss = []
for i in range(1, 100):
    kmeans = KMeans(n_clusters=i, init='k-means++', random_state=42)
    kmeans.fit(movie_features_scaled)
    wcss.append(kmeans.inertia_)
```

```python
# Plotting the Elbow method graph
plt.plot(range(1, 100), wcss)
plt.title('Elbow Method')
plt.xlabel('Number of Clusters')
plt.ylabel('WCSS')
plt.xticks(range(0, 101, 5))
plt.xlim(0, 100)
plt.show()
```

```python
# Found out the elbow curve to be steepest at 37
k = 22
kmeans = KMeans(n_clusters=k, init='k-means++', random_state=42)
kmeans.fit(movie_features_scaled)

# Adding cluster labels to the original dataset
data['cluster'] = kmeans.labels_
```

```python
# Creating a function for similar movie recommendation
def recommend_similar_movies(movie_name, data, kmeans_model, scaler, top_n=3):
    mv_index = data[data['title'] == movie_name].index.values[0]
    mv_features = data.drop(columns=['title', 'cluster']).iloc[mv_index].values.reshape(1, -1)
    mv_scaled = scaler.transform(mv_features)
    predicted_cluster = kmeans_model.predict(mv_scaled)[0]
    cluster_movies = data[data['cluster'] == predicted_cluster]
    similarities = cosine_similarity(movie_features_scaled, scaler.transform(cluster_movies.drop(columns=['title', 'cluster'])))
    similar_movie_indices = similarities.argsort()[0][-top_n-1:-1][::-1]
    similar_movies = cluster_movies.iloc[similar_movie_indices]['title'].tolist()
    return similar_movies
```

```python
movie_name = input()
similar_movies = recommend_similar_movies(movie_name, data, kmeans, scaler)
print("Similar movies to", movie_name, ":")
for movie in similar_movies:
    print("-", movie)
```

## Hardcoded part

```python
import numpy as np
import matplotlib.pyplot as plt
from sklearn.preprocessing import StandardScaler
from sklearn.metrics.pairwise import cosine_similarity
from sklearn.decomposition import PCA
from sklearn.metrics import silhouette_score, calinski_harabasz_score, davies_bouldin_score
from sklearn.metrics import homogeneity_score, completeness_score, v_measure_score
import pickle

# Custom K-means implementation
class CustomKMeans:
    def __init__(self, n_clusters=8, max_iter=300, random_state=None):
        self.n_clusters = n_clusters
        self.max_iter = max_iter
        self.random_state = random_state
        self.centroids = None
        self.labels_ = None
        self.inertia_ = None

    def _initialize_centroids(self, X):
        """Initialize centroids using k-means++ method"""
        np.random.seed(self.random_state)
        n_samples, n_features = X.shape

        # Choose the first centroid randomly
        centroids = np.zeros((self.n_clusters, n_features))
        first_centroid_idx = np.random.randint(0, n_samples)
        centroids[0] = X[first_centroid_idx]

        # Choose the remaining centroids
        for k in range(1, self.n_clusters):
            # Calculate distances from points to the centroids
            distances = np.sqrt(((X - centroids[:k, np.newaxis])**2).sum(axis=2))
            # Distances to closest centroid
            closest_dist = distances.min(axis=0)
            # Probability proportional to squared distance
            weights = closest_dist**2
            # Choose next centroid
            next_centroid_idx = np.random.choice(n_samples, p=weights/weights.sum())
            centroids[k] = X[next_centroid_idx]

        return centroids

    def _get_closest_centroid(self, X, centroids):
        """Assign each sample to the closest centroid"""
        distances = np.sqrt(((X - centroids[:, np.newaxis])**2).sum(axis=2))
        return np.argmin(distances, axis=0)

    def _get_centroids(self, X, labels):
        """Update centroids based on mean of points in each cluster"""
        centroids = np.zeros((self.n_clusters, X.shape[1]))
        for k in range(self.n_clusters):
            if np.sum(labels == k) > 0:  # Avoid empty clusters
                centroids[k] = np.mean(X[labels == k], axis=0)
        return centroids

    def _compute_inertia(self, X, labels, centroids):
        """Calculate sum of squared distances of samples to their closest centroid"""
        distances = np.sqrt(((X - centroids[labels])**2).sum(axis=1))
        return np.sum(distances**2)

    def fit(self, X):
        """Fit K-means to the data"""
        # Initialize centroids
        self.centroids = self._initialize_centroids(X)

        # Main K-means loop
        for i in range(self.max_iter):
            # Assign samples to closest centroids
            old_labels = self.labels_ if self.labels_ is not None else None
            self.labels_ = self._get_closest_centroid(X, self.centroids)

            # Update centroids
            self.centroids = self._get_centroids(X, self.labels_)

            # Check for convergence
            if old_labels is not None and np.all(old_labels == self.labels_):
                break

        # Calculate final inertia
        self.inertia_ = self._compute_inertia(X, self.labels_, self.centroids)
        return self

    def predict(self, X):
        """Predict the closest cluster for each sample in X"""
        return self._get_closest_centroid(X, self.centroids)

# Main process for movie recommendation system
def create_movie_recommendation_system(data, meta=None):
    # Extracting the movie title and features
    movie_titles = data['title']
    feature_columns = [col for col in data.columns if col != 'title']
    movie_features = data[feature_columns]

    # Scaling the features
    scaler = StandardScaler()
    movie_features_scaled = scaler.fit_transform(movie_features)

    # Determining the best k-values using elbow method
    wcss = []
    k_range = range(1, 100)
    for i in k_range:
        kmeans = CustomKMeans(n_clusters=i, random_state=42)
        kmeans.fit(movie_features_scaled)
        wcss.append(kmeans.inertia_)

    # Plotting the Elbow method graph
    plt.figure(figsize=(10, 6))
    plt.plot(k_range, wcss)
    plt.title('Elbow Method')
    plt.xlabel('Number of Clusters')
    plt.ylabel('WCSS')
    plt.xticks(range(0, 101, 5))
    plt.xlim(0, 100)
    plt.grid(True)
    plt.savefig('elbow_method.png')
    plt.close()

    # Using k=34 as mentioned in the code
    k = 34
    kmeans = CustomKMeans(n_clusters=k, random_state=42)
    kmeans.fit(movie_features_scaled)

    # Adding cluster labels to the original dataset
    data['cluster'] = kmeans.labels_

    # Saving the model and necessary components
    model_components = {
        'kmeans': kmeans,
        'scaler': scaler,
        'data': data,
        'movie_features_scaled': movie_features_scaled,
        'feature_columns': feature_columns  # Save the feature columns used
    }

    with open('movie_recommendation_model.pkl', 'wb') as f:
        pickle.dump(model_components, f)

    # If meta data is provided, calculate evaluation metrics
    if meta is not None:
        evaluate_clustering(movie_features_scaled, kmeans.labels_, meta)

    # Visualize clusters using PCA
    visualize_clusters(movie_features_scaled, kmeans.labels_)

    return model_components

def evaluate_clustering(features, labels, meta):
    # Getting True Labels
    true_labels = meta['title'].tolist()

    # Calculating the silhouette score
    silhouette_avg = silhouette_score(features, labels)
    print("Silhouette Score:", silhouette_avg)

    # Calculating the Calinski-Harabasz index
    ch_score = calinski_harabasz_score(features, labels)
    print("Calinski-Harabasz Index:", ch_score)

    # Calculating the Davies-Bouldin index
    db_score = davies_bouldin_score(features, labels)
    print("Davies-Bouldin Index:", db_score)

    # Calculating homogeneity, completeness, and V-measure
    homogeneity = homogeneity_score(true_labels, labels)
    completeness = completeness_score(true_labels, labels)
    v_measure = v_measure_score(true_labels, labels)

    print("Homogeneity:", homogeneity)
    print("Completeness:", completeness)
    print("V-measure:", v_measure)

def visualize_clusters(features, labels):
    # Reducing the feature dimensions to 2 for visualization
    pca = PCA(n_components=2)
    features_2d = pca.fit_transform(features)

    # Plotting the clusters
    plt.figure(figsize=(10, 8))

    # Scatter plot for each cluster
    for cluster_label in set(labels):
        cluster_mask = (labels == cluster_label)
        plt.scatter(features_2d[cluster_mask, 0], features_2d[cluster_mask, 1],
                   label=f'Cluster {cluster_label}', alpha=0.7)

    plt.title('Clusters of Movies (Custom K-means)')
    plt.xlabel('Principal Component 1')
    plt.ylabel('Principal Component 2')
    plt.legend(bbox_to_anchor=(1.05, 1), loc='upper left',
              borderaxespad=0., title='Clusters', ncol=2)
    plt.savefig('cluster_visualization.png')
    plt.close()

def recommend_similar_movies(movie_name, model_components, top_n=3):
    # Unpack model components
    kmeans = model_components['kmeans']
    scaler = model_components['scaler']
    data = model_components['data']
    movie_features_scaled = model_components['movie_features_scaled']

    # Find the movie index
    try:
        mv_index = data[data['title'] == movie_name].index.values[0]
    except IndexError:
        print(f"Movie '{movie_name}' not found in the dataset.")
        return []

    # Instead of dynamically determining feature columns here,
    # we should use the exact same features that were used during training

    # SOLUTION 1: If movie_features_scaled was created with a specific set of columns:
    # Get the original columns used for scaling during model creation
    if 'feature_columns' in model_components:
        feature_columns = model_components['feature_columns']
    else:
        # As a fallback, use what's available but print a warning
        feature_columns = [col for col in data.columns if col not in ['title', 'cluster']]
        print("Warning: Using dynamically determined features which may not match training features")

    # Get movie features
    mv_features = data.loc[mv_index, feature_columns].values.reshape(1, -1)

    # Debug: Check feature dimensions
    print(f"Model expects {scaler.n_features_in_} features, providing {mv_features.shape[1]}")

    # Make sure we have the same number of features as during training
    if mv_features.shape[1] != scaler.n_features_in_:
        print(f"Feature mismatch: Model expects {scaler.n_features_in_} features but got {mv_features.shape[1]}")
        print("Missing features may include: ")
        # This would help identify which feature is missing
        # But would require knowing the original training features
        return []

    mv_scaled = scaler.transform(mv_features)

    # Predict the cluster
    predicted_cluster = kmeans.predict(mv_scaled)[0]

    # Get movies in the same cluster
    cluster_movies = data[data['cluster'] == predicted_cluster]

    # Calculate similarities, but maintain the same feature set
    cluster_features = cluster_movies[feature_columns].values
    cluster_features_scaled = scaler.transform(cluster_features)
    similarities = cosine_similarity(mv_scaled, cluster_features_scaled)[0]

    # Sort by similarity and get top_n (exclude the movie itself)
    similar_indices = np.argsort(similarities)[-top_n-1:-1][::-1]

    # Handle case where there might not be enough movies in the cluster
    if len(similar_indices) < top_n:
        print(f"Warning: Only found {len(similar_indices)} similar movies in cluster {predicted_cluster}")

    similar_movies = cluster_movies.iloc[similar_indices]['title'].tolist()

    return similar_movies

# Example usage
def example_usage():
    # Load the saved model
    with open('movie_recommendation_model.pkl', 'rb') as f:
        model_components = pickle.load(f)

    # Get user input for movie name
    movie_name = input("Enter a movie name: ")

    # Get recommendations
    similar_movies = recommend_similar_movies(movie_name, model_components)

    # Print recommendations
    print(f"Similar movies to '{movie_name}':")
    for movie in similar_movies:
        print("-", movie)
```

```python
import numpy as np
from sklearn.base import BaseEstimator, ClusterMixin
from sklearn.decomposition import PCA
from sklearn.metrics.pairwise import euclidean_distances
import matplotlib.pyplot as plt
from sklearn.metrics.pairwise import cosine_similarity
import pickle

class CFNode:
    """
    Clustering Feature (CF) Node for BIRCH algorithm
    """
    def __init__(self, threshold, branching_factor, is_leaf=True):
        self.threshold = threshold
        self.branching_factor = branching_factor
        self.is_leaf = is_leaf
        self.entries = []  # CF entries or child nodes
        self.n_entries = 0
        self.parent = None

class CFEntry:
    """
    Clustering Feature (CF) Entry for BIRCH algorithm
    """
    def __init__(self, linear_sum=None, squared_sum=None, n=0, child_node=None):
        self.n = n  # Number of points in the cluster
        self.linear_sum = linear_sum  # Sum of all points
        self.squared_sum = squared_sum  # Sum of squares of all points
        self.child_node = child_node  # Points to a child node if not a leaf entry

    def update(self, point):
        """Update the CF entry with a new point"""
        if self.n == 0:
            self.linear_sum = point.copy()
            self.squared_sum = point ** 2
        else:
            self.linear_sum += point
            self.squared_sum += point ** 2
        self.n += 1

    def merge(self, other_entry):
        """Merge with another CF entry"""
        self.n += other_entry.n
        self.linear_sum += other_entry.linear_sum
        self.squared_sum += other_entry.squared_sum

    def get_centroid(self):
        """Calculate the centroid of the CF entry"""
        if self.n > 0:
            return self.linear_sum / self.n
        return None

    def radius(self):
        """
        Calculate the radius (average distance from centroid to all points)
        """
        if self.n <= 1:
            return 0.0

        centroid = self.get_centroid()
        # Using the formula: sqrt((sum(x^2) - 2*centroid*sum(x) + n*centroid^2) / n)
        squared_distance = np.sum(self.squared_sum) - 2 * np.sum(centroid * self.linear_sum) + self.n * np.sum(centroid ** 2)
        return np.sqrt(max(0, squared_distance / self.n))

    def distance(self, other_entry):
        """Calculate the distance between centroids of two CF entries"""
        centroid1 = self.get_centroid()
        centroid2 = other_entry.get_centroid()
        if centroid1 is None or centroid2 is None:
            return float('inf')
        return np.sqrt(np.sum((centroid1 - centroid2) ** 2))

class HardcodedBIRCH(BaseEstimator, ClusterMixin):
    """
    Hardcoded implementation of BIRCH algorithm
    """
    def __init__(self, threshold=0.5, branching_factor=50, n_clusters=None):
        self.threshold = threshold
        self.branching_factor = branching_factor
        self.n_clusters = n_clusters
        self.root = None
        self.labels_ = None
        self.leaf_entries = []

    def fit(self, X):
        """
        Build the CF tree for the dataset X
        """
        self.root = CFNode(self.threshold, self.branching_factor, is_leaf=True)

        # Insert each point into the CF tree
        for i, point in enumerate(X):
            self._insert_point(point)

            # If the root is split, create a new root
            if self.root.n_entries > self.branching_factor:
                old_root = self.root
                self.root = CFNode(self.threshold, self.branching_factor, is_leaf=False)
                self.root.entries.append(self._create_cf_entry_for_node(old_root))
                self.root.n_entries = 1
                old_root.parent = self.root
                self._split_node(old_root)

        # Collect all leaf entries for final clustering
        self.leaf_entries = []
        self._collect_leaf_entries(self.root)

        # Assign cluster labels
        self._assign_labels(X)

        return self

    def _insert_point(self, point):
        """Insert a point into the CF tree"""
        node = self.root

        while not node.is_leaf:
            # Find the closest CF entry
            closest_entry_idx = self._find_closest_entry(node, point)
            node = node.entries[closest_entry_idx].child_node

        # Find the closest leaf entry if any
        if node.n_entries > 0:
            closest_entry_idx = self._find_closest_entry(node, point)
            closest_entry = node.entries[closest_entry_idx]

            # Check if the point can be absorbed without exceeding threshold
            centroid = closest_entry.get_centroid()
            dist = np.sqrt(np.sum((centroid - point) ** 2))

            if dist <= self.threshold:
                # Update the closest entry
                closest_entry.update(point)
                return

        # Create a new entry
        new_entry = CFEntry(linear_sum=point.copy(), squared_sum=point ** 2, n=1)
        node.entries.append(new_entry)
        node.n_entries += 1

        # Check if node needs splitting
        if node.n_entries > self.branching_factor:
            self._split_node(node)

    def _find_closest_entry(self, node, point):
        """Find the index of the closest CF entry to the point"""
        min_dist = float('inf')
        closest_idx = 0

        for i, entry in enumerate(node.entries):
            centroid = entry.get_centroid()
            dist = np.sqrt(np.sum((centroid - point) ** 2))

            if dist < min_dist:
                min_dist = dist
                closest_idx = i

        return closest_idx

    def _create_cf_entry_for_node(self, node):
        """Create a CF entry that summarizes a node"""
        cf_entry = CFEntry(
            linear_sum=np.zeros_like(node.entries[0].linear_sum),
            squared_sum=np.zeros_like(node.entries[0].squared_sum),
            n=0
        )

        for entry in node.entries:
            cf_entry.n += entry.n
            cf_entry.linear_sum += entry.linear_sum
            cf_entry.squared_sum += entry.squared_sum

        cf_entry.child_node = node
        return cf_entry

    def _split_node(self, node):
        """Split a node when it exceeds the branching factor"""
        # Find the two entries that are farthest apart
        max_dist = -1
        farthest_pair = (0, 1)

        for i in range(node.n_entries):
            for j in range(i + 1, node.n_entries):
                dist = node.entries[i].distance(node.entries[j])
                if dist > max_dist:
                    max_dist = dist
                    farthest_pair = (i, j)

        # Create two new nodes
        if node.is_leaf:
            new_node1 = CFNode(self.threshold, self.branching_factor, is_leaf=True)
            new_node2 = CFNode(self.threshold, self.branching_factor, is_leaf=True)
        else:
            new_node1 = CFNode(self.threshold, self.branching_factor, is_leaf=False)
            new_node2 = CFNode(self.threshold, self.branching_factor, is_leaf=False)

        # Add the farthest pair as seeds
        new_node1.entries.append(node.entries[farthest_pair[0]])
        new_node1.n_entries = 1
        new_node2.entries.append(node.entries[farthest_pair[1]])
        new_node2.n_entries = 1

        if not node.is_leaf:
            node.entries[farthest_pair[0]].child_node.parent = new_node1
            node.entries[farthest_pair[1]].child_node.parent = new_node2

        # Distribute the remaining entries
        for i in range(node.n_entries):
            if i != farthest_pair[0] and i != farthest_pair[1]:
                entry = node.entries[i]
                dist1 = entry.distance(new_node1.entries[0])
                dist2 = entry.distance(new_node2.entries[0])

                if dist1 < dist2:
                    new_node1.entries.append(entry)
                    new_node1.n_entries += 1
                    if not node.is_leaf:
                        entry.child_node.parent = new_node1
                else:
                    new_node2.entries.append(entry)
                    new_node2.n_entries += 1
                    if not node.is_leaf:
                        entry.child_node.parent = new_node2

        # Update the parent node
        parent = node.parent
        # Fix: Check if parent exists (it won't for the root node)
        if parent is not None:
            # Find the entry in the parent that points to this node
            for i, entry in enumerate(parent.entries):
                if entry.child_node == node:
                    parent.entries.pop(i)
                    parent.n_entries -= 1
                    break

            cf_entry1 = self._create_cf_entry_for_node(new_node1)
            cf_entry2 = self._create_cf_entry_for_node(new_node2)

            parent.entries.append(cf_entry1)
            parent.entries.append(cf_entry2)
            parent.n_entries += 2

            new_node1.parent = parent
            new_node2.parent = parent

    def _collect_leaf_entries(self, node):
        """Collect all leaf entries for final clustering"""
        if node.is_leaf:
            self.leaf_entries.extend(node.entries)
        else:
            for entry in node.entries:
                if entry.child_node:
                    self._collect_leaf_entries(entry.child_node)

    def _assign_labels(self, X):
        """Assign cluster labels to data points"""
        n_samples = X.shape[0]
        self.labels_ = np.zeros(n_samples, dtype=np.int32)

        if self.n_clusters is None:
            # Use leaf entries directly as clusters
            for j, point in enumerate(X):
                closest_entry_idx = -1
                min_dist = float('inf')

                for k, leaf_entry in enumerate(self.leaf_entries):
                    leaf_centroid = leaf_entry.get_centroid()
                    dist = np.sqrt(np.sum((leaf_centroid - point) ** 2))

                    if dist < min_dist:
                        min_dist = dist
                        closest_entry_idx = k

                self.labels_[j] = closest_entry_idx
        else:
            # Apply additional clustering (e.g., k-means) on leaf centroids
            from sklearn.cluster import KMeans

            # Extract centroids from leaf entries
            centroids = np.array([entry.get_centroid() for entry in self.leaf_entries])

            # Apply k-means on the centroids
            kmeans = KMeans(n_clusters=self.n_clusters, random_state=0)
            leaf_labels = kmeans.fit_predict(centroids)

            # Map leaf labels to data points
            for i, point in enumerate(X):
                closest_entry_idx = -1
                min_dist = float('inf')

                for j, entry in enumerate(self.leaf_entries):
                    centroid = entry.get_centroid()
                    dist = np.sqrt(np.sum((centroid - point) ** 2))

                    if dist < min_dist:
                        min_dist = dist
                        closest_entry_idx = j

                self.labels_[i] = leaf_labels[closest_entry_idx] if closest_entry_idx != -1 else 0

        return self.labels_

    def save_model(self, filename):
        """Save the model to a file"""
        with open(filename, 'wb') as f:
            pickle.dump(self, f)

    @staticmethod
    def load_model(filename):
        """Load the model from a file"""
        with open(filename, 'rb') as f:
            return pickle.load(f)

# Function to recommend similar movies using the custom BIRCH model
def birch_method_movies(movie_name, data, birch_model, scaler, top_n=3):
    movie_index = data[data['title'] == movie_name].index.values[0]
    movie_features = data.drop(columns=['title', 'cluster']).iloc[movie_index].values.reshape(1, -1)
    movie_features_scaled = scaler.transform(movie_features)
    predicted_cluster = birch_model.labels_[movie_index]
    cluster_movies = data[data['cluster'] == predicted_cluster]
    similarities = cosine_similarity(movie_features_scaled, scaler.transform(cluster_movies.drop(columns=['title', 'cluster'])))
    similar_movie_indices = similarities.argsort()[0][-top_n-1:-1][::-1]
    similar_movies = cluster_movies.iloc[similar_movie_indices]['title'].tolist()
    return similar_movies

# Example usage of the custom BIRCH implementation
def main():
    # This is a placeholder for your actual data processing code
    # Replace with your actual data processing code
    from sklearn.preprocessing import StandardScaler
    import pandas as pd

    # Assuming meta is your dataframe
    data = meta.copy()

    # Extracting the movie title and features
    movie_titles = data['title']
    movie_features = data.drop(columns=['title'])

    # Scaling the features
    scaler = StandardScaler()
    movie_features_scaled = scaler.fit_transform(movie_features)

    # Using the custom BIRCH implementation
    birch = HardcodedBIRCH(branching_factor=50, n_clusters=None, threshold=0.5)
    birch.fit(movie_features_scaled)

    # Adding cluster labels to the original dataset
    data['cluster'] = birch.labels_

    # Save the model for later use
    birch.save_model('birch_model.pkl')

    # Visualization using PCA
    pca = PCA(n_components=2)
    movie_features_2d = pca.fit_transform(movie_features_scaled)

    # Plotting the clusters
    plt.figure(figsize=(10, 8))

    # Scatter plot for each cluster
    for cluster_label in set(birch.labels_):
        cluster_mask = (birch.labels_ == cluster_label)
        plt.scatter(movie_features_2d[cluster_mask, 0], movie_features_2d[cluster_mask, 1], label=f'Cluster {cluster_label}')

    plt.title('Clusters of Movies (BIRCH)')
    plt.xlabel('Principal Component 1')
    plt.ylabel('Principal Component 2')
    plt.show()

    # Example of using the model for recommendations
    movie_name = input("Enter a movie name: ")
    similar_movies = birch_method_movies(movie_name, data, birch, scaler)
    print("Similar movies to", movie_name, ":")
    for movie in similar_movies:
        print("-", movie)

if __name__ == "__main__":
    main()
```
